# Choosing a K Value
Let's go ahead and use the elbow method to pick a good K Value!

*Create a for loop that trains various KNN models with different k values, then keep track of the error_rate for each of these models with a list.*

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt

def plot_KNN_error_rate(xdata, ydata):
    error_rate = []
    test_scores = []
    train_scores = []

    ## Split the data into train and test sets in a 70:30 ratio (70% train, 30% test)
    X_train, X_test, y_train, y_test = train_test_split(xdata, ydata, test_size=0.3, random_state=42)

    for i in range(1, 40):
        ## Initialize the KNN classifier with 'i' neighbors
        knn = KNeighborsClassifier(n_neighbors=i)

        ## Fit the KNN model on the training set
        knn.fit(X_train, y_train)

        ## Make predictions on the test set using KNN
        pred_i = knn.predict(X_test)

        error_rate.append(np.mean(pred_i != y_test))  # Error rate calculation
        train_scores.append(knn.score(X_train, y_train))  # Training score
        test_scores.append(knn.score(X_test, y_test))  # Testing score

    # Plotting the error rate vs. K-value
    plt.figure(figsize=(12, 8))
    plt.plot(range(1, 40), error_rate, color='blue', linestyle='dashed', marker='o',
             markerfacecolor='red', markersize=10)
    plt.title('Error Rate vs. K Value')
    plt.xlabel('K')
    plt.ylabel('Error Rate')
    plt.show()

    print()
    # Max training score (training on the same data points)
    max_train_score = max(train_scores)
    train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
    print(f'Max train score {max_train_score * 100} % and k = {list(map(lambda x: x + 1, train_scores_ind))}')

    print()
    # Max test score (testing on the test data points)
    max_test_score = max(test_scores)
    test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
    print(f'Max test score {max_test_score * 100} % and k = {list(map(lambda x: x + 1, test_scores_ind))}')

    return test_scores
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_diabetes
import numpy as np
import matplotlib.pyplot as plt

def plot_KNN_with_minmax_scaling(xdata, ydata):
    # Apply MinMax scaling to the feature data
    scaler = MinMaxScaler()
    xdata_scaled = scaler.fit_transform(xdata)

    # Split the data into train and test sets in a 70:30 ratio (70% train, 30% test)
    X_train, X_test, y_train, y_test = train_test_split(xdata_scaled, ydata, test_size=0.3, random_state=42)

    # Lists to store error rates, training scores, and test scores for each value of k
    error_rate = []
    train_scores = []
    test_scores = []

    # Loop through different k values (1 to 39)
    for i in range(1, 40):
        # Initialize the KNN classifier with 'i' neighbors
        knn = KNeighborsClassifier(n_neighbors=i)

        # Fit the KNN model on the training data
        knn.fit(X_train, y_train)

        # Make predictions on the test data
        pred_i = knn.predict(X_test)

        # Calculate the error rate
        error_rate.append(np.mean(pred_i != y_test))
        # Calculate the training and testing scores
        train_scores.append(knn.score(X_train, y_train))
        test_scores.append(knn.score(X_test, y_test))

    # Plotting the error rate vs. K value
    plt.figure(figsize=(12, 8))
    plt.plot(range(1, 40), error_rate, color='blue', linestyle='dashed', marker='o',
             markerfacecolor='red', markersize=10)
    plt.title('Error Rate vs. K Value (MinMax Scaled Data)')
    plt.xlabel('K')
    plt.ylabel('Error Rate')
    plt.show()

    # Output the max training score and corresponding k value
    max_train_score = max(train_scores)
    train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
    print(f'Max train score {max_train_score * 100} % and k = {list(map(lambda x: x + 1, train_scores_ind))}')

    # Output the max test score and corresponding k value
    max_test_score = max(test_scores)
    test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
    print(f'Max test score {max_test_score * 100} % and k = {list(map(lambda x: x + 1, test_scores_ind))}')

    return test_scores

# Example usage with the sklearn diabetes dataset
diabetes = load_diabetes()
xdata = diabetes.data  # Features (X)
ydata = diabetes.target  # Target (y)

# Call the function to plot the error rate vs. K-value for MinMax-scaled data
plot_KNN_with_minmax_scaling(xdata, ydata)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_diabetes

def plot_voronoi_diagram(xdata, ydata, k=5):
    # Step 1: Apply MinMax scaling to the feature data
    scaler = MinMaxScaler()
    xdata_scaled = scaler.fit_transform(xdata)

    # Step 2: Split the data into train and test sets in a 70:30 ratio (70% train, 30% test)
    X_train, X_test, y_train, y_test = train_test_split(xdata_scaled, ydata, test_size=0.3, random_state=42)

    # Step 3: Initialize the KNN classifier with 'k' neighbors
    knn = KNeighborsClassifier(n_neighbors=k)

    # Step 4: Train the KNN model on the training data
    knn.fit(X_train, y_train)

    # Step 5: Create a meshgrid for plotting the Voronoi diagram
    h = .02  # Step size in the meshgrid
    x_min, x_max = xdata_scaled[:, 0].min() - 1, xdata_scaled[:, 0].max() + 1
    y_min, y_max = xdata_scaled[:, 1].min() - 1, xdata_scaled[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    # Step 6: Make predictions for each point in the meshgrid
    Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Step 7: Plot the Voronoi diagram
    plt.figure(figsize=(10, 8))
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

    # Plot the training points on the diagram
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, s=30, cmap=plt.cm.coolwarm, edgecolors='k', marker='o')
    plt.title(f"Voronoi Diagram for KNN (k={k})")
    plt.xlabel('Feature 1 (scaled)')
    plt.ylabel('Feature 2 (scaled)')
    plt.colorbar()
    plt.show()

# Example usage with the sklearn diabetes dataset
diabetes = load_diabetes()
xdata = diabetes.data[:, :2]  # Select the first two features for 2D visualization
ydata = diabetes.target  # Target variable

# Call the function to plot the Voronoi diagram for KNN with 5 neighbors
plot_voronoi_diagram(xdata, ydata, k=5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# Load the dataset (using sklearn diabetes dataset for this example)
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target

# Step 1: Apply MinMax Scaling to the feature data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Initialize models (KNN, Logistic Regression, and SVM for demonstration)
models = {
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM": SVC(kernel='linear')
}

# Step 3: Apply K-Fold Cross Validation for each model
kf = KFold(n_splits=5, shuffle=True, random_state=42)  # 5-fold cross-validation

# Step 4: Iterate over the models and apply cross-validation
for model_name, model in models.items():
    # Compute cross-validation scores for each fold
    cv_scores = cross_val_score(model, X_scaled, y, cv=kf)

    # Print out the results
    print(f"Model: {model_name}")
    print(f"Cross-validation scores: {cv_scores}")
    print(f"Mean Accuracy: {cv_scores.mean() * 100:.2f}%")
    print(f"Standard Deviation: {cv_scores.std() * 100:.2f}%")
    print("-" * 50)

    # Plotting the cross-validation results
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, len(cv_scores) + 1), cv_scores, marker='o', label=model_name)
    plt.title(f"Cross-Validation Results for {model_name}")
    plt.xlabel('Fold')
    plt.ylabel('Accuracy')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()
```
